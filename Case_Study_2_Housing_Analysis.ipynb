# Exploratory Data Analysis

```python
#Checking the number of married and unmarried in the data
counts = df['marital'].value_counts()
counts
plt.bar(counts.index,counts.values)
plt.title('Martial Status of people')
for i in range(len(counts)):
  plt.text(counts.index[i],counts.values[i],str(counts.values[i]),ha='center',va='top')
plt.show()
```

```python
#Checking the number of peole that own a house
counts_houses=df['housing'].value_counts()
counts_houses
plt.bar(counts_houses.index,counts_houses.values)
plt.title('People owning houses')
for i in range(len(counts_houses)):
  plt.text(counts_houses.index[i],counts_houses.values[i],str(counts_houses.values[i]),ha='center',va='top')
plt.show()
```

```python
#Checking the number people who have loans
counts_loans=df['loan'].value_counts()
counts_loans
plt.bar(counts_loans.index,counts_loans.values)
plt.title('Number of People having loans')
for i in range(len(counts_loans)):
  plt.text(counts_loans.index[i],counts_loans.values[i],str(counts_loans.values[i]),ha='center',va='top')
plt.show()
```

# Further Analysis

```python
#Number of people who are married and have a house
marr_house=df[(df['marital']=='married') & (df['housing']=='yes')]
print('Number of people who are married and have a house:',marr_house.shape[0])

#Number of people who are not married and have a house
not_marr_house=df[(df['marital']=='single') & (df['housing']=='yes')]
print('Number of people who are not married and have a house:',not_marr_house.shape[0])

#Number of people who are married and don't have a house.
marr_no_house=df[(df['marital']=='married') & (df['housing']=='no')]
print("Number of people who are married and don't have a house:",marr_no_house.shape[0])

#Number of people who are not married and don't have a house.
not_marr_no_house=df[(df['marital']=='single') & (df['housing']=='no')]
print("Number of people who are not married and don't have a house:",not_marr_no_house.shape[0])
```

```python
# Extracting the values from the dataframes
marr_house_val = marr_house.shape[0]
not_marr_house_val = not_marr_house.shape[0]
marr_no_house_val = marr_no_house.shape[0]
not_marr_no_house_val = not_marr_no_house.shape[0]

# Creating a bar plot
plt.bar(['Married with House', 'Not Married with House', 'Married without House', 'Not Married without House'],
        [marr_house_val, not_marr_house_val, marr_no_house_val, not_marr_no_house_val])

# Adding title and labels
plt.title('Housing Status by Marital Status')
plt.xlabel('Marital Status and Housing Status')
plt.ylabel('Number of People')
plt.xticks(rotation=90)
for i, v in enumerate([marr_house_val, not_marr_house_val, marr_no_house_val, not_marr_no_house_val]):
    plt.text(i, v+1, str(v), ha='center',va='top')

# Displaying the plot
plt.show()
```
