# Exploratory Data Analysis

```python
# Getting data of the top 20 countries with internet speed in 2021
new_df = pd.DataFrame(df['2021'].sort_values(ascending=False).head(20))
top_countries = pd.DataFrame(df.loc[new_df.index, 'Country Name'].head(20))
print('The top internet speeds are: ',new_df)
print("The top 20 countries are: ",top_countries)
```

```python
#Visualization
plt.figure(figsize=(15,6))
plt.bar(top_countries['Country Name'],new_df['2021'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Getting the top 20 countries by internet speed
new_df_1 = pd.DataFrame(df['2021'].sort_values(ascending=True).head(20))
top_countries_1 = pd.DataFrame(df.loc[new_df_1.index, 'Country Name'].head(20))
print('The bottom internet speeds are: ',new_df_1)
print("The bottom 20 countries are: ",top_countries_1)
```

```python
#Visualization
plt.figure(figsize=(15,6))
plt.bar(top_countries_1['Country Name'],new_df_1['2021'])
plt.xticks(rotation=90)
plt.show()
```
